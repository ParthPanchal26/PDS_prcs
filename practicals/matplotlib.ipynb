```python
import matplotlib.pyplot as plt
```

```python
tech = ['Angular', 'React', 'Vue', '.net', 'Ruby on Rails']
pop = [90, 87, 89, 88, 78]

plt.bar(tech, pop, color='lightblue')
plt.plot(tech, pop, color='orange')
plt.scatter(tech, pop, color='indigo')
plt.title('Tech Popularity')
plt.xlabel('Technologies')
plt.ylabel('Popularity')
plt.show()
```
